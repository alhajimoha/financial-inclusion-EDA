## Exploratory Data Analysis of Financial Inclusion in Africa

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Train.csv")
```

```python
print("Data shape: ", df.shape)
```

```python
df.head()
```

```python
df.info()
```

```python
#Exploring the data further through univariate data analysis
df["bank_account"].value_counts()
```

```python
sns.catplot(x="bank_account", kind="count", data=df)
```

```python
sns.catplot(x="cellphone_access", kind="count", data = df)
```

The data reveals that a majority of our respondents have access to cellphone

```python
sns.catplot(x="education_level", kind="count", data=df)
plt.xticks( rotation = 45, horizontalalignment = "right" )
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.catplot(x="job_type", kind="count", data=df)
plt.xticks(rotation=60, horizontalalignment="right")
plt.show()
```

```python
sns.catplot(x="marital_status", kind="count", data=df)
plt.xticks(rotation=55, horizontalalignment="right")
plt.show()
```

```python
sns.catplot(x="location_type", kind="count", data=df)
plt.show()
```

From this dataset, it shows that more people live in the rural area that urban area

```python
plt.figure(figsize=(10, 6))
df.household_size.hist()
plt.xlabel( "Household Size")
plt.show()
```

The most common individuals living in an household are two and it is not normally distributed


```python
plt.figure(figsize=(10, 6))
df.age_of_respondent.hist()
plt.xlabel("Age of respondent")
plt.show()
```

Most of our respondent are between age 25 to 35

```python
#Diving into Bivariate Analysis
plt.figure(figsize=(10, 6))
sns.countplot(x="location_type", hue="bank_account", data=df)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x="gender_of_respondent", hue= "bank_account", data=df)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x="job_type", hue="bank_account", data=df)
plt.xticks(rotation=55, horizontalalignment="right")
plt.show()
```
